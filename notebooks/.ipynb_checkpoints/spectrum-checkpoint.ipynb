## Beamline spectrum analysis

For some experiments it might be important to measure the spectrum before the experiment, from that we can have an estimative of the bandwidth that is important for certain indexing methods.

Here you can calculate the plot the beamline spectrum from a file containing the intensity measured for each energy. From the peak FWHM one can estimate the bandwidth for the experiment.

import pandas as pd
import matplotlib.pyplot as plt

```python
file_path="/path/to/spectrum.txt"
```

```python
df=pd.read_csv(file_path,sep="  ", names=["energy","intensity"], engine="python")
```

```python
norm_intensity=df.intensity/df.intensity.max()
```

```python
maximum=df.energy.loc[df.intensity.idxmax()]
```

```python
maximum
```

```python
energy=df.energy
```

```python
plt.close()
%matplotlib widget
plt.plot(energy, norm_intensity)
plt.scatter(maximum, 1, color="r")
plt.xlabel("Energy (keV)")
plt.ylabel("Normalized Intensity")
plt.grid()
```

```python
bandwidth = (15.07-14.78)/maximum
```

```python
bandwidth*100 ## in percentage
```
